```python
import numpy as np
import pandas as pd
```

```python
file_path = "National Assembly .csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.describe()
```

```python
df = df.iloc[0:266]
df.columns = df.columns.str.replace(' ', '_')
```

```python
print("Column Name\t\tTotal Null Values")
for column_name in df.columns:
    null_count = df[column_name].isnull().sum()
    print(f"{column_name}\t\t{null_count}")
```

```python
# Convert relevant columns to numeric
numeric_columns = ['Number_of_Voters_Polled_', 'Total_Voters_Registered_']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Check for missing values in 'Number_of_Votes_Polled_', 'Number_of_Voters_Polled_', and 'Total_Voters_Registered_'
missing_votes_polled = df[df['Number_of_Votes_Polled'].isnull()]
missing_voters_polled = df[df['Number_of_Voters_Polled_'].isnull()]
missing_total_registered = df[df['Total_Voters_Registered_'].isnull()]

# Check if there are missing values in 'Number_of_Votes_Polled_' and non-missing values in 'Number_of_Voters_Polled_' and 'Total_Voters_Registered_'
if not missing_votes_polled.empty and not missing_voters_polled.empty and not missing_total_registered.empty:
    # Calculate turnout percentage
    df['Turnout_Percentage'] = df['Number_of_Voters_Polled_'] / df['Total_Voters_Registered_']
    
    # Compute 'Number_of_Votes_Polled_' using the turnout percentage
    df.loc[missing_votes_polled.index, 'Number_of_Votes_Polled'] = df.loc[missing_votes_polled.index, 'Total_Voters_Registered_'] * df.loc[missing_votes_polled.index, 'Turnout_Percentage']
    
    # Drop the 'Turnout_Percentage' column (optional)
    df.drop(columns=['Turnout_Percentage'], inplace=True)

# Check if missing values in 'Number_of_Votes_Polled_' are filled
if df['Number_of_Votes_Polled'].isnull().sum() == 0:
    print("All missing values in 'Number_of_Votes_Polled' have been filled.")
else:
    print("Some missing values in 'Number_of_Votes_Polled' could not be filled.")

# Save the updated dataset
df.to_csv("updated_dataset.csv", index=False)
```
